# Imports

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from gensim import corpora
from gensim.models import LdaModel
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import spacy

pyLDAvis.gensim_models

nlp = spacy.load("en_core_web_sm")
```

# Load Data

```python
df = pd.read_csv("../data/servey.tsv", sep="\t")
df.head()
```

# Analysis

```python
adj_count = dict()
for idx, row in df.iterrows():
    doc = nlp(row[-1])
    for token in doc:
        if token.pos_ == "ADJ":
            lemma = token.lemma_
            if lemma in adj_count:
                adj_count[lemma] += 1
            else:
                adj_count[lemma] = 1
```

```python
top_adj = []
for adj in adj_count:
    max_counts = sorted(adj_count.values())[-10:]
    if adj_count[adj] in max_counts:
        top_adj.append((adj, adj_count[adj]))
```

```python
top_adj = sorted(top_adj, key=lambda x: x[1])
print("\n".join([f"{a}, {b}" for a, b in top_adj]))
```

```python
wordcloud = WordCloud(width=800, height=400).generate_from_frequencies(adj_count)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(range(len(top_adj)), [tup[1] for tup in top_adj])
ax.set_xticks(range(len(top_adj)))
ax.set_xticklabels([tup[0] for tup in top_adj], rotation=45)
ax.set_title("Top 10 Adjective used in response (lemmatized)")
plt.show()
```

```python
analyzer = SentimentIntensityAnalyzer()

sentiment_scores = []
for idx, row in df.iterrows():
    score = analyzer.polarity_scores(row[-1])
    sentiment_scores.append(score["compound"])
```

```python
sentiment = sns.kdeplot(sentiment_scores, fill=True)
plt.xlabel('Sentiment Score')
plt.ylabel('Density')
plt.title('Sentiment Score Density Curve')
plt.show()
```

```python
tokens = []
for idx, row in df.iterrows():
    doc = nlp(row[-1])
    tokens.append([token.lemma_ for token in doc if not token.is_stop])

dictionary = corpora.Dictionary(tokens)
corpus = [dictionary.doc2bow(token) for token in tokens]
model = LdaModel(corpus, num_topics=5, id2word=dictionary)
```

```python
model.show_topics()
```
